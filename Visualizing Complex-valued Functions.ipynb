### Visualizer function

```python
def visualize_complex_func(f, bounds: list, res: int, title: str):
    
    '''
    f - a function
    bounds - [real_min, real_max, imag_min, imag_min]
    res - resolution of the plot
    title - title of the plot
    '''
    
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib
    
    if len(bounds) != 4:
        raise ValueError('4 elements are needed! Please specify the minimum and maximum values of the real part and the imaginary part!')
        
    if isinstance(title, str):
        pass
    else:
        raise ValueError('Please use a string to specify the title of the plot!')
    
    # Creating the meshgrid for complex numbers
    x = np.linspace(bounds[0], bounds[1], res)
    y = np.linspace(bounds[2], bounds[3], res)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j*Y
    C = f(Z)
    
    fig, (ax0, ax1) = plt.subplots(ncols=2)
    ax0.pcolormesh(x, y, np.angle(C), vmin=-np.pi, vmax=np.pi, cmap='hsv')
    ax1.pcolormesh(x, y, np.abs(C), norm=matplotlib.colors.LogNorm(), cmap='viridis')
    ax0.set_aspect('equal')
    ax1.set_aspect('equal')
    ax0.set_title('Î¸')
    ax1.set_title('r')
    fig.suptitle(title, fontsize=14)
    plt.show()
```

```python
# For the function of f(z) = z
visualize_complex_func(lambda x: x, [-1, 1, -1, 1], 400, 'The visualization of f(z) = z')
```

```python
# For the function of f(z) = sqrt(z^2 + 1)
visualize_complex_func(lambda x: np.sqrt(x**2 + 1), [-3, 3, -3, 3], 400, 'The visualization of f(z) = sqrt(z^2 + 1)')
```
